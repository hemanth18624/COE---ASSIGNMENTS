```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("colorado_motor_vehicle_sales.csv")
data.head()
```

```python
data1 = data.groupby(['year'])['sales'].sum()
```

```python
data1
```

```python
x = data1.index
x
```

```python
y = data1.values
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("Years")
plt.ylabel("Amount of Sales in Dollars")
plt.yticks(range(7500000000,15000000000,1000000000))
plt.title("Yearly Sales")
plt.show()
```

```python
shopping_data = pd.read_csv("shopping_trends.csv")
shopping_data.head()
```

```python
age_data = shopping_data['Age']
age_data
```

```python
bins = [10,20,30,40,50,60,70,80]
plt.hist(age_data, bins=bins, edgecolor='black', color='skyblue', alpha=0.7, rwidth=0.8)
plt.xlabel("Age Range")
plt.ylabel("Frequency Count of Customers")
plt.title("Age Distribution of Customers")
plt.show()
```
